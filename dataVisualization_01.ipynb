```python
# Import the library
import pandas as pd
```

```python
# Load the dataset
#df = pd.read_csv("https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro01/oficina/arquivos/dadosoficina01.txt")
df = pd.read_csv('drinks.txt')
```

```python
# View the first 10
df.head(10)
```

1. What is the mean and standard deviation of beer (beer_servings), liquor (spirit_servings), wine (wine_servings) and total alcohol (total_litres_of_pure_alcohol) consumption?

```python
# Calculate the mean
df_m = df.mean()
df_m
```

```python
# Calculate standard deviations
df_std = df.std()
df_std
```

```python
# Average beer consumption
df_m['beer_servings']
```

```python
# Max liquor consumption value
df['spirit_servings'].max()
```

```python
# Min wine consumption value
df['wine_servings'].min()
```

```python
# Standard deviation of alcohol consumption
df_std['total_litres_of_pure_alcohol']
```

```python
# Some statistical measures
df.describe()
```

2. What is the consumption of liquor in France?

```python
# Data on alcohol consumption in Brazil
df[df['country'] == 'Brazil']
```

```python
# The .values returns an array of dataframe values (as the content is just one value, use the first index[0])
df[df['country'] == 'France']['spirit_servings'].values[0]
```

3. What are the 5 countries where the most wine is consumed?

```python
# sort_values and ascending = False to arrange in descending order
df.sort_values(by='wine_servings', ascending = False).head(5)['country'].values
```

4. In which countries beer consumption is above average?

```python
# Countries whose total alcohol consumption is greater than or equal to 12 liters per person
df[df['total_litres_of_pure_alcohol'] >= 12]
```

```python
# df_m['beer_servings'] is the average beer consumption
df[df['beer_servings'] > df_m['beer_servings']]['country'].values
```

5. Which countries in which total alcohol consumption is higher than Brazil, in order of lowest to highest?

```python
# Total alcohol consumption in Brazil
# The .values returns an array of dataframe values. As the content is just one value, use the first index [0]
alcool_brasil = df[df['country'] == 'Brazil']['total_litres_of_pure_alcohol']
alcool_brasil.values[0]
```

```python
# Countries with higher consumption than Brazil
df_alcool = df[df['total_litres_of_pure_alcohol'] > alcool_brasil.values[0]]
# Sort from lowest to highest
df_alcool.sort_values(by='total_litres_of_pure_alcohol', ascending = True)
```
